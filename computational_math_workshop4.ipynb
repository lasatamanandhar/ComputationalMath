Task-1:

Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show

x – Time taken in seconds

y – Distance reached

Title “Buzzer Game Results”

One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]

And add to your scatter() call the ‘c=colors’ argument like so;

What happens?

```python
import matplotlib.pyplot as plt
import random

# Generate random values
x = [random.randint(0,50) for i in range(10)]
y = [random.randint(0,50) for i in range(10)]

# Colors list
colors = ["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]

# Scatter plot
plt.scatter(x, y, c=colors)

plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")

plt.show()
```

Task 2:

Following data shows the change in population of different animals from 2017 to 2022.

a. Create a pie chart for the population of each year

b. Can we show all the pie chart on the single pane? Write the code.

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]

bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

labels = ["Bears", "Dolphins", "Whales"]

import matplotlib.pyplot as plt

for i in range(len(years)):
    data = [bears[i], dolphins[i], whales[i]]

    plt.pie(data, labels=labels)
    plt.title(f"Animal Population in {years[i]}")
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

for i in range(len(years)):
    plt.subplot(2, 3, i+1)   # 2 rows, 3 columns
    data = [bears[i], dolphins[i], whales[i]]

    plt.pie(data, labels=labels)
    plt.title(str(years[i]))

plt.tight_layout()
plt.show()
```

Construct a pie chart to visually display the favorite fruits of the students in a class based on the

```python
import matplotlib.pyplot as plt

colors = ["Red","Silver","Blue","White","Green"]
values = [123,230,78,193,12]

plt.bar(colors, values)
plt.title("Car Data")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()
```

```python
bar_colors = ["red","silver","blue","white","green"]
plt.bar(colors, values, color=bar_colors)
plt.show()
```

```python
plt.pie(values, labels=colors)
plt.title("Car Color Distribution")
plt.show()
```

```python
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(values, labels=colors, explode=myexplode)
plt.show()
```

```python
plt.pie(values, labels=colors, explode=myexplode, shadow=True)
plt.show()
```

```python
plt.pie(values, labels=colors)
plt.legend(title="Car Colours:")
plt.show()
```
