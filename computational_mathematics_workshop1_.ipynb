EXAMPLE 10.

```python
def fib():
    n = int(input("Enter number of terms: "))

    a = 0
    b = 1
    count = 0

    while count < n:
        print(a, end=" ")
        c = a + b
        a = b
        b = c
        count += 1
fib()
```

```python
def power(x, n):
    result = 1
    count = 0

    while count < n:
        result = result * x
        count += 1

    return result
print (power(2, 3))
```

12.

```python
def SumDigits(n):
    total = 0

    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10

    print(total)
SumDigits (71)
```

13.

```python
def squaredSum(arr):
    total = 0
    i = 0

    while i < len(arr):
        total = total + arr[i] * arr[i]
        i += 1

    return total
print(squaredSum([2,3,4]))
```
