```python
import matplotlib.pyplot as plt
import math

x = []
y = []

i = -4 * math.pi
while i <= 4 * math.pi:
    x.append(i)
    y.append(math.sin(i))
    i += 0.1

plt.plot(x, y)
plt.title("y = sin(x)")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

r = 4
a, b = 2, 2

theta = 0
while theta <= 2*math.pi:
    x.append(r * math.cos(theta) + a)
    y.append(r * math.sin(theta) + b)
    theta += 0.1

plt.plot(x, y)
plt.scatter(a, b)  # center point

plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

a, b = 3, 5
p, q = 2, 2

theta = 0
while theta <= 2*math.pi:
    x.append(a * math.cos(theta) + p)
    y.append(b * math.sin(theta) + q)
    theta += 0.1

plt.plot(x, y)
plt.scatter(p, q)

plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x_vals = []
y_vals = []

p, q = 0, 0
a = 0.5

y = -5
while y <= 5:
    x = 4*a*(y-p)**2 + q
    x_vals.append(x)
    y_vals.append(y)
    y += 0.1

plt.plot(x_vals, y_vals)
plt.grid()
plt.show()
```

```python
# Plot 20 values of x and y from the following functions.
# 𝑓(𝑥)=5𝑥3+2𝑥−1
# 𝑓(𝑥)=−2𝑥3+𝑥2+100
# 𝑓(𝑥)=2𝜋𝑥+20


import matplotlib.pyplot as plt
import math

x = list(range(20))

y1 = [5*i**3 + 2*i - 1 for i in x]
y2 = [-2*i**3 + i**2 + 100 for i in x]
y3 = [2*math.pi*i + 20 for i in x]

plt.plot(x, y1, label="f1")
plt.plot(x, y2, label="f2")
plt.plot(x, y3, label="f3")

plt.legend()
plt.show()
```

```python
# Plot all the graph from above functions on the same pane using subplot.

import matplotlib.pyplot as plt

x = list(range(20))

# Function 1
plt.subplot(3, 1, 1)
plt.plot(x, y1, marker='o')
plt.title("f(x) = 5x³ + 2x - 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# Function 2
plt.subplot(3, 1, 2)
plt.plot(x, y2, marker='x')
plt.title("f(x) = -2x³ + x² + 100")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# Function 3
plt.subplot(3, 1, 3)
plt.plot(x, y3, marker='^')
plt.title("f(x) = 2πx + 20")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# Adjust spacing
plt.tight_layout()

# Show all plots
plt.show()
```

```python
# Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋],
# i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋
import math
import matplotlib.pyplot as plt

x = []
y = []

i = -3 * math.pi
while i <= 3 * math.pi:
    x.append(i)
    y.append(math.cos(i))
    i += 0.1

plt.plot(x, y)
plt.show()
```

```python
# Plot a circle with center at (a, b) and radius r.The values of a=b=2, and r=4
import matplotlib.pyplot as plt
import math

x1, y1 = [], []
x2, y2 = [], []

theta = 0
while theta <= 2*math.pi:
    x1.append(4*math.cos(theta)+2)
    y1.append(4*math.sin(theta)+2)

    x2.append(3*math.cos(theta)+2)
    y2.append(5*math.sin(theta)+2)

    theta += 0.1

plt.plot(x1, y1, label="Circle")
plt.plot(x2, y2, label="Ellipse")

plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```

```python
# Plot a Parabola (𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2 with vertex at (𝑝,𝑞).

import matplotlib.pyplot as plt

x = []
y = []

i = -10
while i <= 10:
    x.append(i)
    y.append(i**2)
    i += 0.1

plt.plot(x, y)
plt.show()
```
